```python
def centripetal(m,v,r):
    F=m*v**2/r
    return(F)
```

```python
R=[]
f=[]
for i in range(1,10,1):
    C=centripetal(2,2,i)
    R.append(i)
    f.append(C)
```

```python
print(f)
```

```python
print(R)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=[10,4])
plt.plot(R,f,'rs--',label='force')
plt.xlabel('radius(m)')
plt.ylabel('force(N)')
plt.title('centripetal Force')
plt.legend()
```
